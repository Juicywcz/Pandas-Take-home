1. (Data Cleaning) You need to complete the cleaning of the Example_Data Sheet 
data according to the cleaning rules

Example_Data Sheet contains the following data cleanup logic:
- Company ID and Company Name should be corresponding if there is one ID corresponding to multiple Names or one Name corresponding to 
multiple IDs, the ID/Name is invalid and needs to be cleaned up.
- Fiscal Year is the year, between 1999 and 2021.
- SIC Code is a four-digit number.
- Trading Currency can only be one of two units, USD or GBP.
- SP / CDS / APD / ARD / ADA is an integer number.

```python
# Company ID and Company Name should be corresponding
id_name={}
for c in range(len(original_data)):
    id_name[str(original_data.loc[c,'Company ID'])]=original_data.loc[c,'Company Name']
id_name
```

```python
# D company needs to be deleted, he corresponds to two IDs
example_data=original_data.loc[~original_data['Company Name'].isin(['D Company'])]
# Fiscal Year is the year, between 1999 and 2021.
example_data1=example_data.loc[example_data['Fiscal Year'].isin(list(range(1999,2022)))]
# SIC Code is a four-digit number
example_data1.index=range(len(example_data1))
drop_index=[]
# Traverse each SIC code to determine whether it is a four-digit number, and whether it is a numerical value
for i in range(len(example_data1)):
    sic=example_data1.loc[i,'SIC Code']
    if not len(str(sic))==4:
        drop_index.append(i)
    if not str(sic).isdigit():
        drop_index.append(i)
example_data2=example_data1.drop(drop_index)

# Trading Currency can only be one of two units, USD or GBP.
example_data3=example_data2.loc[example_data2['Trading Currency'].isin(['USD','GBP'])]
# SP / CDS / APD / ARD / ADA is an integer number.
example_data3.index=range(len(example_data3))
# Remove nulls and convert the data into int
example_data3=example_data3.dropna()
example_data3[['SP','CDS','APD','ARD','ADA']]=example_data3[['SP','CDS','APD','ARD','ADA']].astype(int)
# Data processed by task1
example_data3.index=range(len(example_data3))
example_data3
```

```python
# Process SP CDS APD ARD ADA into a separate sequence
deal_columns=['SP','CDS','APD','ARD','ADA']
new_samples=[]
for col in deal_columns: # Traverse each column and each row that needs to be processed
    for v in example_data3[col]: 
        new_samples.append([col, v]) # The length of the list at this time is 45*5=225
new_df=pd.DataFrame(new_samples, columns=['Metric Name','Value'])
example_data4=pd.concat([example_data3]*len(deal_columns)) # Expand example_data3 by 5 times in order to concatenate 225 rows of data
example_data4.index=range(len(example_data4)) # Reset index
deal_example=pd.concat([example_data4, new_df],axis=1) # Concatenate 225 rows of data and delete the original five columns that need to be processed
deal_example=deal_example.drop(deal_columns, axis=1)
deal_example
```
